```python
import torch
```

```python
x_data=[1.0,2.0,3.0]
y_data=[2.0,4.0,6.0]
```

```python
w=torch.Tensor([1.0])
w.requires_grad=True
```

```python
def forward(x):
    return x*w#w是个Tensor,所以x自动类型转换成Tensor,
#因为W需要梯度,所以运算出的结果也需要梯度
```

```python
def loss(x,y):
    y_pred=forward(x)
    return (y_pred-y)**2#代码在构建计算图
```

```python
print("predict before training",4,forward(4).item())#为什么是item(),往回翻,就能那个找到,从上几个视频中找
```

```python
print(forward(4))
```

sum+=l加入到下面程序的空格内,就会变成张量相加,因为l是张量,然后就会占用内存,

```python
for epoch in range(100):
    for x,y in zip(x_data,y_data):
        l=loss(x,y)#计算loss
        l.backward()#方向计算各个梯度,然后是释放计算图
        print('\tgrad',x,y,w.grad.item())#\t是tab的意思,item,是把tensor里边的值去出来变成标量
        w.data=w.data-0.01*w.grad.data#因为grad也是个计算图,所以一定要取到grad的计算图,不会建立计算图,有Tensor会建立计算图
        
        #纯数值得修改,值更新权重
        w.grad.data.zero_()#不清零,梯度会累加
    print('process',epoch,l.item())
print('predict after training',4,forward(4).item())          
```
