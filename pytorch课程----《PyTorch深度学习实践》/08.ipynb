#加载数据两个类
dataset :构造数据集,支持索引,dataloder两个类:拿出min-batch

epoch:定义是所有的训练样本都执行一次就叫一个epoch,一次叫前馈和反馈

batch-size:一次训练用到的样本数量,一次训练之前馈和反馈

iteration:batch一共分了多少个,等于epoch/batch-size

shuffle:打乱顺序

dataloader:知道索引和长度,就可以对dataset自动的小批量的生成

先打乱顺序,然后分组,形成min-batch,iteration-loader

```python
import torch 
import numpy as np
from torch.utils.data import Dataset#Dataset类是抽象类,只能继承不能实例化
from torch.utils.data import DataLoader#这个类是加载数据的
```

```python
class DiabetesDataset(Dataset):
    def __init__(self,filepath):#
        xy=np.loadtxt(filepath,delimiter=',',dtype=np.float32)
        self.len=xy.shape[0]#shape返回(N,9).shape[0]表示N
        self.x_data=torch.from_numpy(xy[:,:-1])
        self.y_data=torch.from_numpy(xy[:,[-1]])
    def __getitem__(self,index):
        return self.x_data[index],self.y_data[index]#返回元组形式
    def __len__(self):
        return self.len
dataset=DiabetesDataset('data/diabetes.csv.gz')
train_loader=DataLoader(dataset=dataset,batch_size=32, shuffle=True, num_workers=0)#需要几个并行的去读数据
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.Linear1=torch.nn.Linear(8,6)
        self.Linear2=torch.nn.Linear(6,4)
        self.Linear3=torch.nn.Linear(4,1)
        self.activate=torch.nn.Sigmoid()
    def forward(self,x):
        x=self.activate(self.Linear1(x))
        x=self.activate(self.Linear2(x))
        x=self.activate(self.Linear3(x))
        return x
model=Model()
criterion=torch.nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)    
```

```python
#if __name__=='__main__':#封装到if与距离不然再window下会报错
for epoch in range(100):#epoch
    for i,data in enumerate(train_loader,0):#x,y放到data
        inputs,labels=data
        y_pred=model(inputs)
        loss=criterion(y_pred,labels)
        print(epoch,i,loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
